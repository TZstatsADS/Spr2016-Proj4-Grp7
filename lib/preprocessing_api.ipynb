```python
import csv

def trycast(x):
    try:
        return float(x)
    except:
        try:
            return int(x)
        except:
            return x

cols = ['product_productid', 'review_userid', 
        'review_helpfulness', 'review_score', 
        'review_time']

f = open('moviescsv.csv', 'wb')
w = csv.writer(f)
w.writerow(cols)


doc =  {}

with open('movies.txt') as infile:
    for line in infile:
        line = line.strip()
        if line=="":
            w.writerow([doc.get(col) for col in cols])
            doc = {}
        else:
            idx = line.find(':')
            key, value = tuple([line[:idx], line[idx+1:]])
            key = key.strip().replace("/", "_").lower()
            value = value.strip()
            doc[key] = trycast(value)
    f.close()
```

```python
rp = data.pivot_table(columns=['product_productid'],index=['review_userid'],values='review_score')
#rp.head()
```

```python
rp = rp.fillna(0)
```

```python
Q = rp.values
```

```python
Q.shape
```

```python
W = Q>0.5
W[W == True] = 1
W[W == False] = 0
# To be consistent with our Q matrix
W = W.astype(np.float64, copy=False)
```

```python
lambda_ = 0.1
n_factors = 100
m, n = Q.shape
n_iterations = 20
```

```python
X = 5 * np.random.rand(m, n_factors) 
Y = 5 * np.random.rand(n_factors, n)
```

```python
def get_error(Q, X, Y, W):
    return np.sum((W * (Q - np.dot(X, Y)))**2)
```

```python
weighted_errors = []
for ii in range(n_iterations):
    for u, Wu in enumerate(W):
        X[u] = np.linalg.solve(np.dot(Y, np.dot(np.diag(Wu), Y.T)) + lambda_ * np.eye(n_factors),
                               np.dot(Y, np.dot(np.diag(Wu), Q[u].T))).T
    for i, Wi in enumerate(W.T):
        Y[:,i] = np.linalg.solve(np.dot(X.T, np.dot(np.diag(Wi), X)) + lambda_ * np.eye(n_factors),
                                 np.dot(X.T, np.dot(np.diag(Wi), Q[:, i])))
    weighted_errors.append(get_error(Q, X, Y, W))
    print('{}th iteration is completed'.format(ii))
weighted_Q_hat = np.dot(X,Y)
#print('Error of rated movies: {}'.format(get_error(Q, X, Y, W)))
```

# Collaborative filtering approach （Alternating Least Square implementation）
##### (http://bugra.github.io/work/notes/2014-04-19/alternating-least-squares-method-for-collaborative-filtering/)
#### Intuition: people who agreed in the past will agree in the future, and that they will like similar kinds of items as they liked in the past.

#### Advantages: 
##### 1, No need to know about item content
##### 2, Capture user interest
#### Disadvantages:
##### 1, Sparsity of user preference (66345 users and 409 movies 110000 reviews in total)
##### 2, Scalability in computation (110000 users and 30000 movies 6500000 reviews)
